6. olist_products_dataset.csv:

product_id: unique identifier for each product

product_category_name: name of the category that the product belongs to

product_name_lenght: number of characters in the product name

product_photos_qty: number of photos for the product

product_weight_g: weight of the product in grams

7. olist_sellers_dataset.csv:

seller_id: unique identifier for each seller

seller_zip_code_prefix: zip code prefix for the seller's location

seller_city: city where the seller is located

seller_state: state where the seller is located

8. product_category_name_translation.csv:

 ### Cheaking for duplicates

```python
# A function to check for duplicates in our datasets
def check_duplicates(df):
    """
    This function checks for and returns any duplicates in a given dataframe.
    """
    duplicates = df[df.duplicated()]
    if duplicates.shape[0] == 0:
        print("No duplicates found in the dataset")
    else:
        print("Duplicates found in the dataset:")
        return duplicates
```

```python
# Calling for the function to check for duplicates
check_duplicates(customers)
check_duplicates(geolocation)
check_duplicates(items)
check_duplicates(payment)
check_duplicates(reviews)
check_duplicates(orders)
check_duplicates(products)
check_duplicates(sellers)
check_duplicates(product_name)
```

```python
# A function to check for outliers
def check_outliers(data):
    for col in data.select_dtypes(include='number').columns:
        plt.boxplot(data[col])
        plt.title(col)
        plt.show()
```

# cheaking for missing values

```python
# A function to check for missing values in our dataset
def check_missing_values(data):
    # Count missing values in each column
    missing_values = data.isnull().sum()

    # Convert missing values count to percentage of total rows
    missing_percent = (missing_values / len(data)) * 100

    # Combine the missing values count and percent into a DataFrame
    missing_df = pd.concat([missing_values, missing_percent], axis=1)
    missing_df.columns = ['Missing Values', '% of Total']

    # Return only columns with missing values
    missing_df = missing_df[missing_df['Missing Values'] > 0]

    return missing_df

# Check missing values in each dataset
display(check_missing_values(customers))
display(check_missing_values(geolocation))
display(check_missing_values(items))
display(check_missing_values(payment))
display(check_missing_values(reviews))
display(check_missing_values(orders))
display(check_missing_values(products))
display(check_missing_values(sellers))
display(check_missing_values(product_name))
```
